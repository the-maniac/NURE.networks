Creating Graph


```python
import matplotlib.pyplot as plt
import networkx as nx
import helpers


N = 8192
L = 1.25
nodes = list(range(2))
G = nx.Graph([nodes], name="Test Graph", label=True)
node = 2
while node < N:
    num_of_edges = helpers.get_number_of_edges(L, G)
    probabilities = helpers.get_nodes_probabilities(G)
    rnd_nodes = helpers.get_random_nodes(G, probabilities, num_of_edges)
    G.add_edges_from(zip([node] * num_of_edges, rnd_nodes))
    node += 1
```

Degree distribution

```python
import matplotlib.pyplot as plt
degrees = dict(G.degree())
degree_values = sorted(set(degrees.values()))
histogram = [list(degrees.values()).count(i)/float(nx.number_of_nodes(G)) for i in degree_values]

plt.figure()
plt.bar(degree_values, histogram)
plt.title('Degree Distribution of Test Graph')
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.show()
```

Degree distribution

```python
import matplotlib.pyplot as plt
degrees = dict(G.degree())
degree_values = sorted(set(degrees.values()))
histogram = [list(degrees.values()).count(i)/float(nx.number_of_nodes(G)) for i in degree_values]

plt.figure()
plt.plot(degree_values, histogram, 'o')
plt.title('Degree Distribution of Test Graph')
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

Rank degree distribution

```python
import matplotlib.pyplot as plt

degree_sequence = sorted([d for n, d in G.degree()], reverse=True)

dmax = max(degree_sequence)

plt.loglog(degree_sequence, 'b-', marker='o')

plt.title("Degree rank distribution of Test Graph")
plt.ylabel("degree")
plt.xlabel("rank")
plt.show()
```

Drawing graph

```python
import matplotlib.pyplot as plt

nx.draw(G, with_labels=False, font_weight='bold', node_size=1)

plt.show()
```
